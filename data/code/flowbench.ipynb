```python
import numpy as np
import matplotlib.pyplot as plt
from matplotlib.animation import FuncAnimation
from IPython.display import HTML

%matplotlib inline
```

```python
case = '0000'
```

```python
type = ['bc', 'geo', 'prop']

path = f'CFDBench/cylinder/{type[0]}/case{case}/v.npy'
```

```python
data = np.load(path)
```

```python
data.shape
```

```python
"""
visualize
"""
fig, ax = plt.subplots(figsize=(10, 6))

# Use actual data range for color scaling
vmin, vmax = data.min(), data.max()

im = ax.imshow(data[0], cmap='RdBu_r', animated=True, 
               vmin=vmin, vmax=vmax, origin='lower', aspect='auto')
cbar = plt.colorbar(im, ax=ax, label='u-velocity')
ax.set_title(f'Cylinder Flow - Frame 0/{len(data)-1}')
ax.set_xlabel('X')
ax.set_ylabel('Y')

def update(frame):
    im.set_array(data[frame])
    ax.set_title(f'Cylinder Flow - Frame {frame}/{len(data)-1}')
    return [im]

# Animate all 620 frames
anim = FuncAnimation(fig, update, frames=620, 
                     interval=50, blit=True)

# To save: anim.save('cylinder_flow.gif', writer='pillow', fps=20)
plt.close()
HTML(anim.to_html5_video())
```
